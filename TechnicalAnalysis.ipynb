```python

import yfinance as yf
import plotly.express as px
import pandas as pd
import numpy as np
np.set_printoptions(precision=3, linewidth=200)
from plotly.subplots import make_subplots
```

```python

# Define the ticker symbol and the date range
ticker_symbol = 'PG'  # P&G ticker symbol
end_date = pd.to_datetime('today').strftime('%Y-%m-%d')
start_date = (pd.to_datetime('today') - pd.DateOffset(years=3)).strftime('%Y-%m-%d')

# Retrieve historical data
stock_data = yf.download(ticker_symbol, start=start_date, end=end_date)
```

```python
dates=stock_data.index.values
orig_closing = stock_data.loc[:,'Close'].values.squeeze()
percent_changes=orig_closing[1:]/orig_closing[:-1]
idx=np.arange(percent_changes.shape[0])
x=np.random.uniform(0,1,size=(20,percent_changes.shape[0]))
rand_idxs=np.argsort(x, axis=1)
fakes=orig_closing[0]*np.cumprod(percent_changes[rand_idxs],axis=1)
```

```python
fig=make_subplots(rows=5,cols=4)
for r in range(5):
    for c in range(4):
        fig.add_scatter(x=dates, y=fakes[r*4+c,:], row=r+1, col=c+1)
fig.update_layout(showlegend=False, template='plotly_dark')
```
