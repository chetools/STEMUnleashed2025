```python

import yfinance as yf
import plotly.express as px
import pandas as pd
import numpy as np
np.set_printoptions(precision=3, linewidth=200)
from plotly.subplots import make_subplots
```

```python

# Define the ticker symbol and the date range
ticker_symbol = 'PG'  # P&G ticker symbol
end_date = pd.to_datetime('today').strftime('%Y-%m-%d')
start_date = (pd.to_datetime('today') - pd.DateOffset(years=3)).strftime('%Y-%m-%d')

# Retrieve historical data
stock_data = yf.download(ticker_symbol, start=start_date, end=end_date)
```

```python
dates=stock_data.index.values
orig_closing = stock_data.loc[:,'Close'].values.squeeze()
percent_changes=orig_closing[1:]/orig_closing[:-1]
idx=np.arange(percent_changes.shape[0])
x=np.random.uniform(0,1,size=(20,percent_changes.shape[0]))
rand_idxs=np.argsort(x, axis=1)
fakes=orig_closing[0]*np.cumprod(percent_changes[rand_idxs],axis=1)
```

```python
fig=make_subplots(rows=5,cols=4)
for r in range(5):
    for c in range(4):
        fig.add_scatter(x=dates, y=fakes[r*4+c,:], row=r+1, col=c+1)
fig.update_layout(width=800, height=800, showlegend=False, template='plotly_dark')
```

```python
x= np.linspace(-1, 1, 20)
y = 2*x + 1
y = y + np.random.normal(0, 0.2, size=x.size)
```

```python
A = np.c_[x, np.ones_like(x)]
c = np.linalg.solve(A.T @ A, A.T @ y)
```

```python
fig=make_subplots()
fig.add_scatter(x=x, y=y, mode='markers')
fig.add_scatter(x=x, y=A @ c, mode='lines')
fig.update_layout(width=600, height=400, showlegend=False, template='plotly_dark')
```

```python
def sse(p):
    m,b = p
    return np.sum(((m*x+b)-y)**2)
```

```python
sse(c)
```

```python
from scipy.optimize import minimize
```

```python
minimize(sse, np.array([1.8, 0.8]))
```

```python
A=np.c_[np.sin(x), np.sin(1.5*x)]
```

```python
# error = y - A@c
# A.T @ error = 0  : error is perpendicular to each of the basis vectors
# A.T @ (y - A@c)
# A.T @ y - A.T @ A @ c = 0 : distributive property
# A.T @ A @ c = A.T @ y : equation to solve
```

```python
c = np.linalg.solve(A.T @ A, A.T @ y)
```

```python
x=np.linspace(0,6*np.pi,100)
y=2.5*np.sin(x)+1.2*np.sin(1.5*x)
fig=make_subplots()
fig.add_scatter(x=x, y=y, mode='markers')
fig.add_scatter(x=x, y=A @ c, mode='lines')
fig.update_layout(width=600, height=400, showlegend=False, template='plotly_dark')
```

```python
x_orig=np.linspace(0,10*np.pi,100)
y_orig=2.5*np.sin(x_orig)+0.3*x
fig=make_subplots()
fig.add_scatter(x=x_orig, y=y_orig, mode='markers')
fig.update_layout(width=600, height=400, showlegend=False, template='plotly_dark')
```

```python
m,b = np.polyfit(x_orig, y_orig, 1)
fig=make_subplots()
fig.add_scatter(x=x_orig, y=y_orig, mode='markers')
fig.add_scatter(x=x_orig, y=m*x_orig+ b, mode='lines')
fig.update_layout(width=600, height=400, showlegend=False, template='plotly_dark')
```

```python
idxs = np.where(y_orig<m*x_orig+b)[0]
x,y = x_orig[idxs],y_orig[idxs]
m,b = np.polyfit(x, y, 1)
fig=make_subplots()
fig.add_scatter(x=x_orig, y=y_orig, mode='markers')
fig.add_scatter(x=x, y=y, mode='markers')
fig.add_scatter(x=x, y=m*x+ b, mode='lines')
fig.update_layout(width=600, height=400, showlegend=False, template='plotly_dark')
```
